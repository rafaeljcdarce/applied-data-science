```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200 # 200 e.g. is really fine, but slower
```

```python
df = pd.read_csv('/Users/rileyevans/Downloads/recent.csv')
df
```

```python
pos_df = df[df['sents'] == 1][['t', 'c']]
```

```python
neg_df = df[df['sents'] == -1][['t', 'c']]
```

```python
mask = np.array(Image.open('/Users/rileyevans/Downloads/452-4528141_solid-color-map-of-usa-hd-png-download.png.png'))
```

```python
def plot(text):
    wc = WordCloud(background_color="white", max_words=1000, mask=mask)
    # generate word cloud
    wc.generate_from_frequencies(text)

    # show
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.savefig('recent-hashtag-neg-wordcloud.png')
    plt.show()
```

```python
plot(neg_df.set_index('t').to_dict()['c'])
```

```python
plot(pos_df.set_index('t').to_dict()['c'])
```
